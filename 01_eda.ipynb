## 3. Target Variable Analysis (Hedef Değişken Analizi)
#### Class Distribution (Sınıf Dağılımı)
- **Legitimate transactions**: 283,253 (99.833%)
- **Fraudulent transactions**: 473 (0.167%)
### Imbalance ratio: 598.8:1
- Her 598 legitimate işlemde sadece 1 fraud var
- Bu **extreme imbalanced dataset**
### Implications for Modeling (Modelleme için Çıkarımlar)
1. **Accuracy yanıltıcı olacak**: Model her şeyi "0" tahmin etse %99.8 accuracy alır!
2. **Metric seçimi kritik**: Accuracy yerine **ROC-AUC, Precision, Recall, F1-Score** kullanmalıyız
3. **Class weighting gerekli**: Model eğitirken `class_weight='balanced'` kullanmalıyız dolandırıcıya odaklanmak için
4. **Sampling stratejileri**:Undersampling, oversampling düşünebiliriz
5. **Threshold optimization**: Default 0.5 yerine optimal threshold bulmalıyız
6. **Scaling (Ölçeklendirme):** Time ve Amount sütunlarını normalize etmemiz gerekecek. (PCA yapılmış V sütunları ile Amount sütunu arasındaki uçurumu kapatmak için).

### Business Perspective (İş Perspektifi)
- Fraud detection'da **False Negative (fraud'u kaçırmak)** çok maliyetli!
- False Positive (legitimate'i fraud sanmak) de müşteri deneyimini bozar
- Balance bulmalıyız: **Recall** (fraud'ları yakala) vs **Precision** (false alarm'ı azalt)
---

### Metric Seçiminin Önemi
1.**Recall (Duyarlılık):** "Gerçek hırsızların kaçını yakaladım?" sorusunun cevabıdır. Banka için en önemlisi budur. Hırsızı kaçırmak (False Negative) bankaya milyonlar kaybettirir.

2.**Precision (Kesinlik):** "Hırsız dediğim kişilerin kaçı gerçekten hırsız?" sorusudur. Eğer herkese hırsız dersek müşterilerle iletişim iyi olmaz (False Positive).

3.**F1-Score**: Bu ikisinin ortalamasıdır. Dengeli bir başarı için buna bakılır.

4.**ROC-AUC:** Modelin 0 ve 1'i birbirinden ne kadar iyi ayırdığını gösteren genel bir puandır.

```python
print("korelasyon analizi")
korelasyon_target = df.corr()["Class"].sort_values(ascending=False)
print("en yüksek korelasyon olanlar")
print(korelasyon_target.head(15))
print("en düşük korelasyona sahip olanlar")
print(korelasyon_target.tail(15))
```

```python
# Top 15 pozitif + top 15 negatif
top_features = korelasyon_target.drop('Class').head(15)
bottom_features = korelasyon_target.drop('Class').tail(15)

# Birleştir
important_features = pd.concat([top_features, bottom_features])

# Bar chart
plt.figure(figsize=(12, 6))
colors = ['green' if x > 0 else 'red' for x in important_features.values]
plt.barh(important_features.index, important_features.values, color=colors)
plt.xlabel('Sınıf ile korelasyon (Fraud)')
plt.title('Fraud ile özellik korelasyonu', fontsize=14, fontweight='bold')
plt.axvline(x=0, color='black', linestyle='--', linewidth=1)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Top 10 önemli feature seç
top_10_features = korelasyon_target.drop('Class').abs().sort_values(ascending=False).head(10).index

# Bu feature'lar + Class için heatmap
selected_features = list(top_10_features) + ['Class']
correlation_matrix = df[selected_features].corr()

# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Top 10 Features - Korelasyon Matrix', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print("\nTop 10 En Önemli Feature'lar:")
for i, feature in enumerate(top_10_features, 1):
    corr_value = korelasyon_target[feature]
    print(f"{i}. {feature}: {corr_value:.3f}")
```
